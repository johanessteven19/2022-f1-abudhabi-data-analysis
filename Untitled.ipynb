```python
import pandas as pd

lap_times = pd.read_csv('lap_times.csv')
drivers = pd.read_csv('drivers.csv')

abudhabi = lap_times.loc[lap_times['raceId'] == 1096]

abudhabi
```

```python
match_drivers = drivers[drivers['driverId'].isin(abudhabi['driverId'])][[ 'forename', 'surname', 'number', 'code', 'driverId' ]]
match_drivers = match_drivers.merge(abudhabi[['driverId', 'lap', 'position', 'time', 'milliseconds']], on='driverId')
abudhabi2022 = match_drivers.sort_values(['lap', 'position'], ascending=True)
abudhabi2022[:10]
```

```python
from matplotlib import pyplot as plt
import seaborn as sns

# ax.set(ylim=(1:31.000, 1:35.000))
# sns.set(rc={'figure.figsize':(25,15)})
sns.lineplot(data=abudhabi2022, x='lap', y='milliseconds', hue='surname')
```
