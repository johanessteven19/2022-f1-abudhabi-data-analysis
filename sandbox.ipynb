```python
import pandas as pd

lap_times = pd.read_csv('lap_times.csv')
drivers = pd.read_csv('drivers.csv')

abudhabi = lap_times.loc[lap_times['raceId'] == 1096]

abudhabi
```

```python
match_drivers = drivers[drivers['driverId'].isin(abudhabi['driverId'])][[ 'forename', 'surname', 'number', 'code', 'driverId' ]]
match_drivers = match_drivers.merge(abudhabi[['driverId', 'lap', 'position', 'time', 'milliseconds']], on='driverId')
abudhabi2022 = match_drivers.sort_values(['lap', 'position'], ascending=True)
abudhabi2022[:10]
```

```python
perez_laptimes = abudhabi2022.loc[abudhabi2022['forename'] == 'Sergio'][['surname', 'lap', 'milliseconds']]
# abudhabi = lap_times.loc[lap_times['raceId'] == 1096]
perez_laptimes
```

```python
leclerc_laptimes = abudhabi2022.loc[abudhabi2022['forename'] == 'Charles'][['surname', 'lap', 'milliseconds']]
# abudhabi = lap_times.loc[lap_times['raceId'] == 1096]
leclerc_laptimes
```

```python
import numpy as np
delta_laptimes_perlec = np.empty([1,0])
counter = 0
while counter < 58:
    delta_laptimes_perlec = np.append(delta_laptimes_perlec, (perez_laptimes['milliseconds'].iloc[counter] - leclerc_laptimes['milliseconds'].iloc[counter]))
    counter+=1

delta_laptimes_perlec

# # leclerc_laptimes['milliseconds'].iloc[0] - perez_laptimes['milliseconds'].iloc[0]
```

```python
delta_lap = pd.DataFrame(delta_laptimes_perlec)
```

```python
delta_lap['lap'] = 1
x = 0
while x < 58:
    delta_lap['lap'][x] = x+1
    x += 1
delta_lap
```

```python
delta_lap = delta_lap.rename(columns={0: "delta"})
```

```python
# ax.set(ylim=(1:31.000, 1:35.000))
sns.set(rc={'figure.figsize':(25,15)})
sns.barplot(data=delta_lap, x="lap", y="delta")
```
